```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Assuming 'filtered_df' is your DataFrame
differential_pressure_array = np.array(filtered_df['differential pressure value'])
sample_rate = 12  

def find_positive_peaks(signal):
    
    signal = signal - np.mean(signal) # Removing the mean value of the signal 
    peaks = []
    positive_part = []
    
    for i, value in enumerate(signal):
        if value > 0:
            positive_part.append(value)
        elif positive_part:
            
            peak_index = positive_part.index(max(positive_part))
            peaks.append(peak_index + i - len(positive_part))
            positive_part = []  # Reset positive_part for the next positive part
    
    # Check for any remaining positive part at the end of the signal
    if positive_part:
        peak_index = positive_part.index(max(positive_part))
        peaks.append(peak_index + i - len(positive_part) + 1)
    
    return peaks

# Plot the original signal
time = np.arange(0, len(differential_pressure_array)) / sample_rate
plt.plot(time, differential_pressure_array, label='Original Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Differential Pressure Signal Plot')
plt.legend()
plt.show()

# Find and plot the peaks
peaks = find_positive_peaks(differential_pressure_array)


plt.plot(time, differential_pressure_array, label='Original Signal')
plt.plot(time[peaks], differential_pressure_array[peaks], 'r.', label='Peaks')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Differential Pressure Signal with Peaks')
plt.legend()
plt.show()

# Display the indices of the peaks
print("Indices of Peaks:", peaks)
```
