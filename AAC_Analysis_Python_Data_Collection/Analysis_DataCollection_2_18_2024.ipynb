```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
from scipy.signal import find_peaks
```

```python
import pandas as pd

# Read the text file
                   
df = pd.read_csv('../AAB_DataCollection/DataCollection_2_18_2024/DataCollection_2_18_2024.txt')  # replace '\t' with your delimiter

# Save the DataFrame as a CSV file
df.to_csv('../AAB_DataCollection/DataCollection_2_14_2024/DataCollection_2_18_2024.csv', index=False)

df.head()  # prints the first 5 rows of the DataFrame
```

```python
df.columns
```

```python
import numpy as np

df['differential pressure value'] = np.where(df['Data'].str.strip() == 'Differential Pressure values:', df['Data'].shift(-1), np.nan)
df['pressure values'] = np.where(df['Data'].str.strip() == 'Pressure values:', df['Data'].shift(-1), np.nan)

# Fill NaN values
df['differential pressure value'].fillna(method='ffill', inplace=True)
df['pressure values'].fillna(method='ffill', inplace=True)

df.head(100)
```

```python
df.head(100)  # prints the first 5 rows of the DataFrame
```

```python

# Assuming df is your DataFrame

# Find rows where any cell has a NaN value
nan_rows = df[df.isna().any(axis=1)]
print(nan_rows)
# Drop rows where any cell has a NaN value
df = df.dropna()
```

```python
# Define start and end time
start_time = pd.to_datetime('16:12:00')
end_time = pd.to_datetime('16:14:00')

# Filter rows between start time and end time
filtered_df = df.copy()
filtered_df = filtered_df[(filtered_df['Timestamp'] >= start_time) & (filtered_df['Timestamp'] <= end_time)]

filtered_df = filtered_df.copy()
# Now you can modify 'pressure values' without warnings
filtered_df.loc[:, 'differential pressure value'] = pd.to_numeric(filtered_df['differential pressure value'], errors='coerce')


# Create a new figure with a specific size (width, height)
plt.figure(figsize=(10, 6))

# Plot 'Data' against 'Timestamp'
plt.plot(filtered_df['Timestamp'], filtered_df['differential pressure value'])

# Set the title and labels
plt.title('Normal Breathing Mouth Closed')
plt.xlabel('Timestamp')
plt.ylabel('Differential Pressure Readings')

# Adjust the spacing of the labels
plt.xticks(rotation=45)

# Set the y-axis limits to cover the entire range
#plt.ylim(float('-5'), float('90'))

# Add a grid
plt.grid(True)

plt.savefig('Calibration_2_1_2024.png')
# Show the plot
plt.show()

############################################################################################################################

# Assuming 'filtered_df' is your DataFrame
differential_pressure_array = np.array(filtered_df['differential pressure value'])
sample_rate = 12  

signal = differential_pressure_array - np.mean(differential_pressure_array) 


sign = [0]*len(signal)
signal = np.array(signal)
segments=[]
count = 0
valley_index=[];
# Iterate through the signal
for i, value in enumerate(signal):
    if value < 0:
        if i == len(signal) - 1:
            count += 1
            segments.append(count)
            valley_end, _ = find_peaks(-signal[i-count+1:i+1],distance=count)
            valley_index.extend(valley_end+(i-count+1))
            count = 0

        sign[i] = -1
        count += 1  
    else: 
        #Positive or 0 
        if i != 0 and sign[i - 1] == -1:   
            #Positive but the previous was negative 
            segments.append(count)
            valley, _ = find_peaks(-signal[i-count:i],distance=count)
            valley_index.extend(valley+(i-count))
            count=0
            
        sign[i] = +1    
print(segments)
print(valley_index)


# Plot the original signal
time = np.arange(0, len(signal))/12
plt.plot(time, signal, label='Original Signal')
#plt.plot(time, signal, label='Zero Based Signal
plt.plot(time[valley_index], signal[valley_index], 'r.', label='Peaks')
plt.plot(time,sign, label='Sign')
#draw the valley as x  on the plot 
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Signal + Valleys')
plt.grid()
plt.legend()
plt.show()
```
