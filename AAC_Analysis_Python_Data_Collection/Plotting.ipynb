```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Correct file path
file_path = r'C:\Users\scrpa\OneDrive - Politecnico di Milano\Desktop\America\UIC\THESIS_2\SmartRetainer\AAA_InterfaceProcessing\data_2024-05-07_14-19-13.csv'

# Check if the file exists
print(os.path.exists(file_path))  # Should return True if the file is present

# Reading the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# Convert Timestamp from milliseconds to seconds
df['Timestamp'] = df['Timestamp'] / 1000

# Plotting Pressure_abs against Timestamp
plt.figure(figsize=(10, 6))
plt.plot(df['Timestamp'], df['Pressure_abs'], label='Pressure_abs', color='blue')
plt.xlabel('Timestamp (seconds)')
plt.ylabel('Pressure_abs')
plt.title('Pressure_abs over Time')
plt.legend()
plt.grid(visible=True)
plt.show()
```

```python
# Select data between 0 and 80 seconds
df_filtered = df[(df['Timestamp'] >= 0) & (df['Timestamp'] <= 80)]

# Plotting Pressure_abs against Timestamp
plt.figure(figsize=(10, 6))
plt.plot(df_filtered['Timestamp'], df_filtered['Pressure_abs'], label='Pressure_abs', color='blue')
plt.xlabel('Timestamp (seconds)')
plt.ylabel('Pressure_abs')
plt.title('Test Under Stress')
plt.legend()
plt.grid(visible=True)

# Save the plot as an image
output_image_path = r'C:\Users\scrpa\OneDrive - Politecnico di Milano\Desktop\America\UIC\THESIS_2\SmartRetainer\AAA_InterfaceProcessing\pressure_abs_plot.png'
plt.savefig(output_image_path)
plt.show()
```
